```python
from dataligo import Ligo
```

```python
# path of config file. see the sample config for reference (https://github.com/VinishUchiha/dataligo/blob/main/sample_ligo_config.yaml)
config_path = 'path_of_ligo_config_file'

# Create ligo object
ligo = Ligo(config_path)
```

```python
ligo
```

```python
# Create the gcs ligo object
gcs = ligo.connect('gcs')
```

### Read

```python
# Read a dataframe using gcs_path
df = gcs.read_as_dataframe('gs://dcx_test/folder1/data.csv')
df.head()
```

```python
# Read a dataframe using bucket and blob name
df = gcs.read_as_dataframe(bucket='dcx_test', blob_name='folder2/data.parquet')
df.head()
```

```python
# Read dataframe with additional pandas arguments
df = gcs.read_as_dataframe('gs://dcx_test/folder1/data.csv', pandas_args={'encoding':'latin-1','encoding_errors':'ignore'})
df.head()
```

### Write

```python
# Write a dataframe
gcs.write_dataframe(df, bucket='dcx_test',blob_name='test_write/data.csv')
```

```python
# Write a dataframe with additional arguments
gcs.write_dataframe(df, bucket='dcx_test',blob_name='test_write/data.csv',index=False,sep=',')
```

```python
# Write a dataframe in other extensions
gcs.write_dataframe(df, bucket='dcx_test',blob_name='test_write/data.parquet')
#gcs.write_dataframe(df, bucket='dcx_test',blob_name='test_write/data.xlsx')
gcs.write_dataframe(df, bucket='dcx_test',blob_name='test_write/data.json')
```
